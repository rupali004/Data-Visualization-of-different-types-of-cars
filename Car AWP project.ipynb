# Add Grid Lines to a Plot

```python
plt.figure(figsize=(5,5))
car["price"].hist(bins=10)
plt.title("price frequency distribution")
plt.axvline(car_mean,color="red")
plt.axvline(car_median,color="yellow")
plt.ylabel("frequency")
plt.xlabel("price")
plt.legend(labels=["mean","median"])
plt.show()
```

```python
l=["brand","city","model_name"]
def barplot(l):
    for i in l:
        plt.figure(figsize=(10,10))
        car[i].value_counts().plot(kind='bar')
        plt.title(i+" "+"counts")
        plt.ylabel("Counts")
        plt.xlabel(l)
        plt.xticks(rotation=90)    
        plt.show()
barplot(l)
```

```python
# so if it is right side skewd let calculate mean and median for price 
plt.figure(figsize=(6,6))
sns.distplot(car['price'])
plt.axvline(car_mean,color="red")
plt.axvline(car_median,color="yellow")
plt.show()
```

```python
lscat=["brand","city","fuel_type"]
for i in lscat:
    plt.figure(figsize=(6,6))
    sns.countplot(data=car,x=i)
    plt.xticks(rotation=270) 
    plt.show()
```

# pie chart

```python
plt.figure(figsize=(10,12))
car['brand'].value_counts().plot.pie(autopct="%1.1f%%")  # %1.1 show decimal point after dot
plt.title("Brand count percentage")
plt.show()
```

```python
plt.figure(figsize=(7,7))
car['city'].value_counts().plot.pie(autopct="%1.1f%%,")  # %1.1 show decimal point after dot
plt.title("city count percentage")
plt.show()
```

```python
#explode represents how far from the center each wedge is displayed
# wedge 0.2 from the center of the pie:
```

```python
plt.figure(figsize=(7,7))
car['city'].value_counts().plot.pie(autopct="%1.1f%%",explode=(0.2,0,0,0,))  # %1.1 show decimal point after dot
plt.title("city count percentage")
plt.show()
```
